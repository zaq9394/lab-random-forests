```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
numerical=pd.read_csv(r'C:\Users\Admin\IH-Labs\lab-random-forests\files_for_lab\numerical.csv')
categorical=pd.read_csv(r'C:\Users\Admin\IH-Labs\lab-random-forests\files_for_lab\categorical.csv')
target=pd.read_csv(r'C:\Users\Admin\IH-Labs\lab-random-forests\files_for_lab\target.csv')
```

```python
numerical
```

```python
for i in categorical.columns:
    categorical[i]=categorical[i].astype('object')
```

```python
target.head()
```

```python
data = pd.concat([numerical, categorical, target], axis = 1)
```

```python
y = data['TARGET_B']
X = data.drop(['TARGET_B','TARGET_D'], axis = 1) # TARGET_D
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
numericalX = X_train.select_dtypes(np.number)
categoricalX = X_train.select_dtypes(object)

numericalX_test= X_test.select_dtypes(np.number)
categoricalX_test = X_test.select_dtypes(object)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first', handle_unknown = 'ignore').fit(categoricalX)
encoded_categorical = encoder.transform(categoricalX).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical)

encoded_categorical_test = encoder.transform(categoricalX_test).toarray()
encoded_categorical_test = pd.DataFrame(encoded_categorical_test)
```

```python
from sklearn.preprocessing import MinMaxScaler

transformer = MinMaxScaler().fit(numericalX)
X_norm = transformer.transform(numericalX)
X_train_norm = pd.DataFrame(X_norm, columns=numericalX.columns)
X_train_norm.head()

X_test_norm=pd.DataFrame(transformer.transform(numericalX_test),columns=numericalX_test.columns)
X_test_norm.head()
```

```python
X_train_transformed=pd.concat([encoded_categorical,X_train_norm],axis=1)
X_test_transformed=pd.concat([encoded_categorical_test,X_test_norm],axis=1)
```

```python
y_train = y_train.reset_index(drop=True) 
y_test = y_test.reset_index(drop=True) 
```

```python
# Upsampling
```

```python
train = pd.concat([X_train_transformed, y_train],axis=1)
```

```python
target_b_0 = train[train['TARGET_B']==0]
target_b_1 = train[train['TARGET_B']==1]
```

```python
from sklearn.utils import resample
```

```python
target_b_1_oversampled = resample(target_b_1, 
                                    replace=True,
                                    n_samples = len(target_b_0),
                                    random_state=0)
```

```python
train_oversampled = pd.concat([target_b_0,target_b_1_oversampled],axis=0)
```

```python
train_oversampled = train_oversampled.sample(frac =1)
```

```python
y_train_over = train_oversampled['TARGET_B'].copy()
X_train_over = train_oversampled.drop('TARGET_B',axis = 1).copy()
```

```python
X_train = pd.DataFrame(X_train_over)
X_test = pd.DataFrame(X_test_transformed)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=20,
                             min_samples_leaf =20,
                             max_samples=0.2,
                             random_state = 42)
clf.fit(X_train, y_train_over)
print(clf.score(X_train, y_train_over))
print(clf.score(X_test, y_test))

y_pred = clf.predict(X_test)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```
